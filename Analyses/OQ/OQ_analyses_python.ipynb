```python
# Defining functions
import subprocess
import re
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def analyze(job_ini):
    
    proc = subprocess.run(
        ["oq", "engine", "--run", job_ini],
        capture_output=True,
        text=True
    )
    return proc

def output(proc):
    output = proc.stdout + proc.stderr
    
    match = re.search(r"calc_\d+", output)
    calc_id = match.group(0) #This gets a value like calc_15, making it easy to access the output
    export_folder = f"output/outputs_{calc_id}"
        
    return export_folder

def export(export_folder, output):

#    export_folder = f"output_{calc_id}"
    
    proc2 = subprocess.run(
        ["oq", "engine", "--export-outputs", export_folder[20:len(export_folder)], f"output/{output}"],
        capture_output=True,
        text=True
    )

    for i in os.listdir(f"output/{output}"):
        if i.endswith(".csv"):
#            print(f"output/{export_folder}/{i}")
            df = pd.read_csv(f"output/{output}/{i}")
            df.to_csv(f"output/{output}/{i}", index=False, header=False)
            r = re.search(r'_\d+.csv', i)
            if r != None:
                i.replace(r.group(0), '.csv')
                c = i.replace(r.group(0), '.csv')
                os.rename(f"output/{output}/{i}", f"output/{output}/{c}")
#os.rename(i, c)

#    print(f"output/{output}")
#    print(os.getcwd())
```
